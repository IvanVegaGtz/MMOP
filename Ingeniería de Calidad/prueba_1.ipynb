```python
def get_histo(image, num_bin, plot):
    histo = np.zeros(num_bin)
    image_flatten = image.flatten()
    for pixel in image_flatten:
        indice_bin = int(pixel/256 * num_bin)
        histo[indice_bin] += 1
    if plot:
        plt.plot(range(num_bin), histo[0:num_bin])
    return histo
```

```python
def co_ocurrencia_matriz(img, d, theta):
    M = np.zeros((256,256), dtype = float)
    m, n = img.shape
    for i in range(m):
        for j in range(n):
            if theta == 0:
                try :
                    p1 = int(img[i,j])
                    p2 = int(img[i,j+d])
                    M[p1][p2] += 1
                except IndexError:
                    j = n
            elif theta == 45:
                try:
                    p1 = img[i,j]
                    p2 = img[i+d,j+d]
                    M[p1][p2] += 1
                except IndexError:
                    j = n
            elif theta == 90:
                try:
                    p1 = img[i,j]
                    p2 = img[i+d,j]
                    M[p1][p2] += 1
                except IndexError:
                    j = n
    return M    
```

```python
def sum_hist(img, d, plot):
    H = np.zeros(511)
    m, n = img.shape
    for i in range(m):
        for j in range(n):
            try :
                p1 = int(img[i,j])
                p2 = int(img[i,j+d])
                H[p1+p2] += 1
            except IndexError:
                j = n
    if plot:
        plt.plot(range(len(H)), H[0:len(H)])
    return H
```

```python
def res_hist(img, d, plot):
    H = np.zeros(511)
    m, n = img.shape
    for i in range(m):
        for j in range(n):
            try :
                p1 = int(img[i,j])
                p2 = int(img[i,j+d])
                H[p1 - p2 + 255] += 1
            except IndexError:
                j = n
    if plot:
        plt.plot(range(len(H)), H[0:len(H)])
    return H
```

```python
def get_ft_glcm(G):
    m, n = G.shape
    ent = 0
    con = 0
    hom = 0
    for i in range(m):
        for j in range(n):
            ent -= G[i,j]*math.log(G[i,j] + 1)
            con += G[i,j]*(i - j)**2
            hom += G[i,j]*(1/(1 + (i-j)**2))
    features = [ent, con, hom]
    return features
```

```python
def get_matrix_distance(img, prototipo):
    m, n, chanels  = img.shape
    b = img[:,:,0]
    g = img[:,:,1]
    r = img[:,:,2]
    distance_matrix = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            pixel_blue = b[i][j]
            pixel_green = g[i][j]
            pixel_red = r[i][j]
            distance = (pixel_blue - prototipo[0])**2 + (pixel_green - prototipo[1])**2 + (pixel_red - prototipo[2])**2
            distance_matrix[i][j] = distance**0.5
    return distance_matrix
```
