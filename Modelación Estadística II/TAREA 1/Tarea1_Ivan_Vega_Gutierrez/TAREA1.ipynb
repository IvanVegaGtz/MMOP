## Limpieza de datos

Para iniciar con la limpieza de los datos, se procederá a eliminar las columnas Id y Sinopsis, ya que no proporcionan información útil.

La fecha de lanzamiento de manera específica como viene en la base de datos no es de utilidad, ya que al ser tan particular difícilmente estas fechas tendrán una frecuencia mayor a dos, sin embargo el mes podría ser un dato interesante, ya que de esa manera podríamos agrupar los datos y plantearnos si el mes de lanzamiento influye en las recaudaciones de una película.

Debido a que la columna de Título trae el año en el que se estrenó la película y tiene más observaciones que no son NaN como en el caso de la columna de fecha de lanzamiento, extraemos el año de la columna Título y creamos una nueva variable llamada Año.

Por último, la variable Tiempo, represental el tiempo que dura la película, sin embargo no es de utilidad tal como viene en la base de datos, por lo que se procede a convertir el tiempo en minutos y poder trabajar con datos numéricos.


```python
# Limpieza de datos

# Eliminamos las columnas Id y Sinopsis.
peliculas.drop(['Id', 'Sinopsis'],axis='columns', inplace=True)

# Hallamos el mes en el que se realizó la película
# Eliminamos los caracteres en Lanzamiento
for i in range(10):
    num = str(i)
    peliculas['Fecha'] = peliculas['Fecha'].replace({num:''},regex=True)
# Eliminamos la coma sobrante
peliculas['Fecha'] = peliculas['Fecha'].replace({',':''},regex=True)

#Extraemos el año en el que se estrenó la película
peliculas['Año'] = 0
for i in range(len(peliculas['Título'])):
    peliculas['Año'][i] = peliculas['Título'][i][-5:-1]

#Convertimos el tiempo en unidades de minutos
for i in range(len(peliculas['Tiempo'])):
    tiempo = int(peliculas['Tiempo'][i][0])*60
    if peliculas['Tiempo'][i][-1] == 'n':
        tiempo = tiempo + int(peliculas['Tiempo'][i][5] + peliculas['Tiempo'][i][6])
    peliculas['Tiempo'][i] = tiempo


peliculas
```

En los siguientes bloques de código se guarda la información sobre el tiempo de duración en cuatro intervalos, para poder determinar que películas son las más popoulares.

```python
# Creamos un vector que contenga intervalos de tiempo para poder analizarlo
intervalo = np.zeros(4)
for i in range(len(peliculas['Tiempo'])):
    if peliculas['Tiempo'][i] < 90:
        intervalo[0] = intervalo[0] +1
    elif 90 <= peliculas['Tiempo'][i] < 120:
        intervalo[1] = intervalo[1] + 1
    elif 120 <= peliculas['Tiempo'][i] < 150:
        intervalo[2] = intervalo[2] + 1
    else:
        intervalo[3] = intervalo[3] + 1
        
intervalo
```

```python
times = ['Menos de hora y media', 'Entre 1:30 hrs y 2:00 hrs', 'Entre 2:00 hrs y 2:30 hrs', 'Más de dos horas y media']
plt.rcParams["figure.figsize"] = (10, 5)
sns.barplot(times, intervalo)
plt.title("Tiempos de duración de las películas")
plt.xlabel('Intervalos de tiempo')
plt.ylabel('Número de películas')
```

```python
print("El tiempo promedio qued dura una película es ", peliculas['Tiempo'].mean())
print("La desviación estándar de la duración de las peliculas es ", peliculas['Tiempo'].std())
```

Otra variable de interés es la de Género, sin embargo cada película puede contener más de un género. Lo que nos interesa saber es si existe un género predominante en la base de datos. Por lo tanto, en el siguiente bloque de código se procede a extraer todos los géneros que hay y a marcar con 1 si la película tiene el género y 0 en caso contrario.

```python
# Veamos cuantos generos hay
# Creamos la lista para guardar los generos
Gen = []
for i in range(len(peliculas['Género'])):
    generos = peliculas['Género'][i]
    #Eliminamos caracteres no deseados
    generos = generos.replace("'",'')
    generos = generos.replace("[",'')
    generos = generos.replace(']','')
    generos = generos.replace(' ','')
    lista = generos.split(',')
    for j in range(len(lista)):
        if lista[j] not in Gen:
            Gen.append(lista[j])
print(Gen)
print(len(Gen))

#Creamos columnas con los nombres de los géneros
for i in range(len(Gen)):
    peliculas[Gen[i]] = 0
    for j in range(len(peliculas['Género'])):
        if Gen[i] in peliculas['Género'][j]:
            peliculas[Gen[i]][j] = 1

# Ponemos 1 si la película contiene el género y 0 en caso contrario

#print(type(peliculas['Género'][917]))
peliculas
```

```python
freq_Gen = []
for gen in Gen:
    freq_Gen.append(peliculas[gen].sum())

plt.title('Géneros de películas')
plt.xlabel('Géneros')
plt.ylabel('Número de películas')
plt.rcParams["figure.figsize"] = (10, 6)
sns.barplot(Gen,freq_Gen)
plt.xticks(rotation=45)
```
